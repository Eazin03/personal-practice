```python
import pandas as pd

# 构建DataFrame -- 创建数据集(学员成绩信息)
df = pd.DataFrame([
    {"姓名": "小王","语文": 80, "数学": 90, "英语": 80},
    {"姓名": "小张","语文": 90, "数学": 80, "英语": 90},
    {"姓名": "小李","语文": 80, "数学": 84, "英语": 90},
    {"姓名": "小王1","语文": 91, "数学": 83, "英语": 83},
    {"姓名": "小王2","语文": 81, "数学": 93, "英语": 82},
    {"姓名": "小王3","语文": 92, "数学": 82, "英语": 81},
])

# 统计计算
print(f'语文最高分: {df["语文"].max()}, 语文最低分: {df["语文"].min()}, 语文平均分: {df["语文"].mean():.2f}')
print(f'数学最高分: {df["数学"].max()}, 数学最低分: {df["数学"].min()}, 数学平均分: {df["数学"].mean():.2f}')
print(f'英语最高分: {df["英语"].max()}, 英语最低分: {df["英语"].min()}, 英语平均分: {df["英语"].mean():.2f}')
```
