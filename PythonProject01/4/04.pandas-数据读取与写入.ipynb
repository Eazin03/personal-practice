```python
df3 = pd.read_csv("data/sales_data.csv")

# 配置项,展示所以的数据
# pd.set_option('display.max_rows', None)

# 数据过滤
# 1.获取 销量大于等于10的数据
df3[df3['销售数量'] >= 10]

# 2.获取产品类别为食物 或图书的数据
df3[df3['产品类别'].isin(['食品', '图书'])]

# 3.获取单价在100-200之间的数据 -->between: 两边边界都包含
df3[df3['单价'].between(125,193,inclusive="right")]# inclusive: 设置包含边界 right: 右边界包含在内, left: 左边界包含在内, both: 两边界都包含在内,neither: 两边界都不包含在内

# 4. 获取销售数量 >=8 , 并且单价 >= 100的数据 ---> 多条件(&: 并且, |: 或)
df3[(df3['销售数量'] >= 8) & (df3['单价'] >= 100)] # 多条件要用括号括起来

# 5.获取产品类别为 服装/食品 , 支付方式为微信/支付宝的数据
df3[(df3['产品类别'].isin(['服装', '食品'])) & (df3['支付方式'].isin(['微信支付', '支付宝']))]
```
