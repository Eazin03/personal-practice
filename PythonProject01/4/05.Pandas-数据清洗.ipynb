```python
import pandas as pd

pd.set_option("display.min_rows", 30) #设置显示的最小行数(默认为10行)

# 1.读取数据
df = pd.read_csv('data/sales_data.csv')

# 2.数据清洗
# 2.1 缺失值处理
df.isnull()#检查是否有缺失值

# 2.1.1 删除缺失值
df.dropna()# 删除有缺失值的行
df1 = df.dropna(axis=1)# 删除有缺失值的列

# 2.1.2 填充缺失值
df.fillna("--")# 填充缺失值
df.ffill() # 运用前一个值进行填充
df.bfill() # 运用后一个值进行填充


```

```python
# 2.2 重复值处理
# 2.2.1 查看重复值
df.duplicated() # 检查是否有重复值(默认检查所有列)
df.duplicated(subset=["订单号"]) # 检查指定列是否有重复值


# 2.2.2 删除重复值
df.drop_duplicates(subset=["订单号"],keep="first") # keep='last' 保留最后出现的重复值; keep=False 删除所有重复值 ; keep='first' 保留第一个
```

```python
# 2.3 异常值处理
# 2.3.1 查看异常值
df.loc[df["单价"] < 0]


# 2.3.2 删除异常值
df.drop(df[df["单价"] < 0].index)

# 2.3.3 填充异常值
df["单价"] = df["单价"].abs()# 绝对值

```

```python
# 2.4数据格式处理
# df['订单日期'] = df['订单日期'].replace('2023-01-01', '2023-01-01 00:00:00') # 替换,但不推荐
df['订单日期'] = df['订单日期'].str.replace('/','-')
df
```
