```python
import pandas as pd

# 1.读取数据
df = pd.read_csv("data/sales_data.csv", nrows=10) # nrows=10 只读取10行数据

# 2.排序
# 2.1 根据销售数量倒序排序
df.sort_values('销售数量', ascending=False)

# 2.2 根据销量数量升序排序
df.sort_values('销售数量', ascending=True)

# 2.3 根据多个字段排序
# 在第一个列数据相同,在根据第二个字段排序
df.sort_values(['销售数量', '单价'], ascending=[False, True])
```

```python
df = pd.read_csv("data/sales_data.csv", nrows=20)
df

# 3.分组
# 3.1 根据产品类别分组, 统计各个类别的订单数量 - count --缺失值不计算
df.groupby('产品类别').size()
df.groupby('产品类别')['订单号'].count()

# 3.2 根据产品类别分组, 统计各个类别的销售数量之和  sum --缺失值不计算
df.groupby('产品类别')['销售数量'].sum()

# 3.3 根据产品类别分组, 统计各个类别的销售金额之和  sum
df['销售金额'] = df['单价'] * df['销售数量']
df.groupby('产品类别')['销售金额'].sum()

# 3.4 根据产品类别分组, 统计各个类别的单价平均值  mean
df.groupby('产品类别')['单价'].mean()

# 3.5 根据产品类别分组, 统计各个类别的单价最高 值  max
df.groupby('产品类别')['单价'].max()

# 3.6 根据产品类别分组, 统计各个类别的单价最低 值  min
df.groupby('产品类别')['单价'].min()

# 3.7 根据产品类别分组, 统计各个类别的商品的平均单价 最高单价 最低单价 - agg
df.groupby('产品类别').agg({'单价': ['mean', 'max', 'min']})
df.groupby('产品类别')['单价'].agg(['mean', 'max', 'min'])

# 3.8 根据产品类别分组, 统计各个类别的商品的销量数量之和, 销售金额之和, 平均单价
c = df.groupby('产品类别').agg({'销售数量': 'sum', '销售金额': 'sum', '单价': 'mean'}).to_dict()


```
