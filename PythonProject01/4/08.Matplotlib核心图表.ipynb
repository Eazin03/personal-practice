```python
from matplotlib.axes import Axes
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure

# 展示中文
plt.rcParams['font.sans-serif'] = ['SimHei'] # 设置中文字体为黑体

# 创建子图
# figure: 图像对象 , axes: 子图对象(里面存放的是Axes类型的对象)
figure , axes = plt.subplots(nrows=1, ncols=2,figsize=(20,6),dpi=100) # nrows: 行数, ncols: 列数, figsize: 图像大小, dpi: 图像分辨率

# 图1:柱状图(世界石油储备) --->bar
countries = ["中国", "美国", "日本", "印度", "法国", "英国", "俄罗斯", "印度尼西亚", "德国", "意大利"]
values = [135.0, 100.0, 70.0, 60.0, 50.0, 40.0, 30.0, 20.0, 15.0, 10.0]

axes1:Axes = axes[0] # : Axes 为 Axes 对象,引入是为了告诉系统,axes1 是一个Axes对象
axes1.set_title("世界石油储备" , fontsize=16)
axes1.bar(countries, values, color="green",width=0.6) # 绘制柱状图 参数: x轴数据, y轴数据, 颜色, 宽度
axes1.set_xlabel("国家", fontsize=14) # 设置x轴标签
axes1.set_ylabel("油量(亿桶)", fontsize=14) # 设置y轴标签
axes1.grid(linestyle='--', alpha=0.5) # 绘制网格线


# 图2:饼状图(世界人口) -->pie
countries2 = ["中国", "美国", "日本", "印度", "法国", "英国", "俄罗斯", "印度尼西亚", "德国", "意大利"]
values2 = [135.0, 100.0, 70.0, 60.0, 50.0, 40.0, 30.0, 20.0, 15.0, 10.0]

axes2 :Axes = axes[1]
axes2.pie(values2, labels=countries2, autopct="%1.1f%%", startangle=90)# 绘制饼状图 , 参数: 数据, 标签, 百分比(%1.1f%%: 保留一位小数,两边加上%是为了解析%,小数点前面的1,代表着最少宽度), 起始角度
axes2.set_title("世界人口" , fontsize=16)
axes2.legend(loc="lower center", ncol=5, bbox_to_anchor=(0.5, -0.1))#bbox_to_anchor=(0.5,-0.1): 设置图例的位置,前一个参数是左右,后一个是上下, loc: 设置图例的位置, ncol: 设置图例的列数

# 保存图片
plt.savefig("./data/matplotlib_01.png")


plt.show()# 显示图像 ; 注意:在Jupyter中可以直接显示出来,而在python中需要使用plt.show()显示图像
```
